```python
from ultralytics import YOLO
data_yaml = '/content/fish_dataset/data.yaml'
model = YOLO('yolov8n.pt')
```

```python
import yaml

data_yaml_path = '/content/fish_dataset/data.yaml'

with open(data_yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Update the validation path to the correct Colab path
data['val'] = '/content/fish_dataset/images/val'

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)

print(f"Updated '{data_yaml_path}' with the correct validation path.")

# Update the training path to the correct Colab path
data['train'] = '/content/fish_dataset/images/train'

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)

print(f"Updated '{data_yaml_path}' with the correct training path.")
```

```python
model.train(
    data=data_yaml,
    epochs=100,       # You can start with 10-20 for testing
    imgsz=640,        # Input image size
    batch=16,         # Adjust if CUDA memory error occurs
    device=0          # 0 = GPU
)
```
